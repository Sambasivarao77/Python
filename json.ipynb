JSON is Java Script Object Notation  
JSON is data exchange fomat similar to XML

```python
book = {}
book['tom'] = {
    'name':"tom",
    "address":"tom_address",
    "phone_number":12334567897
}
book['bom'] = {
    'name':"bom",
    "address":"bom_address",
    "phone_number":71233456789
}

#convert this dict to string format to store in file
import json
s = json.dumps(book)
print(s)

#store data in Super.txt
with open("Super.txt","w") as f:
  f.write(s)
```

```python
#retrieve data from Super.txt
f =  open("Super.txt","r")
r = f.read()
print(r)
#convert the string format to dict
book = json.loads(r)
print(book)

#print phone number of tom
print(book["tom"]["phone_number"])
```
