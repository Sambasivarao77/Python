```python
from multiprocessing import Pool

def f(n):
  return n*n

if __name__ == "__main__":
  array = [1,2,3,4,5]
  p = Pool()
  result = p.map(f,array)
  print(result)
```

```python
#measure peroformance 
from multiprocessing import Pool
import time
def f(n):
  sums = 0
  for i in range(10):
    sums += n*n*n*n
    time.sleep(0.01)
  return sums

if __name__ == "__main__":

  t2 = time.time()
  result = []
  for x in range(1000):
    result.append(f(x))
  
  print("Serial processing took : ",time.time()-t2)

  t1 = time.time()
  p = Pool()
  result = p.map(f,range(1000))  # f function will be called for 2L times and each time 1000 iterations => 2L*1000
  p.close()
  p.join()

  print("Pool took time : ",time.time()-t1)
  


```

```python
from multiprocessing import Pool

def f(n):
  return n*n

if __name__ == "__main__":
  p = Pool(processes = 4) # will distribute work to only 4 processes
  result = p.map(f,range(5))
  print(result)
```
