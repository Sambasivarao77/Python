```python
f = open("Super.txt","r")
for line in f:
  print(line)

f.close()
```

```python
f = open("Super.txt","r")
word_count = 0
for line in f:
  tokens = line.split(" ")
  print(str(tokens))
  word_count += len(tokens)
  
print("words count = ",word_count)
f.close()
```
